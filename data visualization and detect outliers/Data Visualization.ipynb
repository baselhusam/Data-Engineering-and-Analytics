```python
Remove outliers

df_jord = df[df['Country'] =='Jordan'].reset_index(drop=True)

Q1 = df_jord.describe()['TotalPrice']['25%']
Q3 = df_jord.describe()['TotalPrice']['75%']
IQR = Q3 - Q1
Upper = Q3 + 1.5*IQR
Lower = Q1 - 1.5*IQR

no_outliers_lower = len(df_jord[df_jord['TotalPrice'] <Lower])
no_outliers_higher = len(df_jord[df_jord['TotalPrice'] >Upper])

cond = (df_jord['TotalPrice'] <Lower) | (df_jord['TotalPrice'] > Upper)

df_outliers_only  =df_jord[cond]

df_without_outliers = df_jord[~cond]
```
