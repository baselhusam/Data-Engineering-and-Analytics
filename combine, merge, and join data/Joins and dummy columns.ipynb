```python
import numpy as np
import pandas as pd

A = pd.DataFrame(np.random.randint(1,10,(3,2)), columns=['Col_1', 'Col_2'])
A['ID'] = A.index.to_list()

B = pd.DataFrame(np.random.randint(1,10,(5,2)), columns=['Col_3', 'Col_4'])
B['ID'] = B.index.to_list()
```

```python
A
```

```python
B
```

```python
pd.concat([A,B], axis=1)
```

```python
pd.concat([A,B], axis=0)
```

```python
pd.concat([A,B], axis=1, join='inner')
```

```python
pd.concat([A,B], axis=1, join='outer')
```

```python
A.merge(B, on='ID', how='inner')
```

```python
A.merge(B, on='ID', how='outer')
```

```python
A.merge(B, on='ID', how='left')
```

```python
A.merge(B, on='ID', how='right')
```
